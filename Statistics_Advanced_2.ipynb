Q1 - What is hypothesis testing in statistics

  Ans - Hypothesis testing is a statistical method used to make decisions or draw conclusions about a population based on sample data.

The basic idea is to test whether there is enough evidence in the sample to support a particular claim.

Steps in Hypothesis Testing

State the hypotheses

Null hypothesis (H₀): The default assumption, usually stating that there is no effect, difference, or relationship.
Alternative hypothesis (H₁ or Hₐ): The claim you want to test, stating that there is an effect, difference, or relationship.

Example:

H₀: The average height of students is 170 cm.
H₁: The average height of students is not 170 cm.
Choose a significance level (α)
Commonly α = 0.05 (5%).
This is the maximum probability of rejecting H₀ when it is actually true.
Collect data and calculate a test statistic
Examples: z-statistic, t-statistic, chi-square statistic, F-statistic.
Find the p-value
The p-value measures how likely the observed data would be if H₀ were true.
Make a decision
If p-value ≤ α, reject H₀.
If p-value > α, fail to reject H₀.

Q2 - What is the null hypothesis, and how does it differ from the alternative
hypothesis?

  ANS - The null hypothesis (H₀) and alternative hypothesis (H₁ or Hₐ) are the two competing statements in a hypothesis test.

Null Hypothesis (H₀)

The null hypothesis represents the default assumption or status quo. It usually states that there is no effect, no difference, or no relationship.

Examples:

A new drug has the same effectiveness as the current drug.
The average exam score is 75.
There is no difference in customer satisfaction between two stores.
Alternative Hypothesis (H₁ or Hₐ)

The alternative hypothesis is the claim that the researcher wants to investigate. It states that there is an effect, difference, or relationship.

Examples:

A new drug is more effective than the current drug.
The average exam score is not 75.
Customer satisfaction differs between the two stores.

| Null Hypothesis (H₀)                            | Alternative Hypothesis (H₁ or Hₐ)               |
| ----------------------------------------------- | ----------------------------------------------- |
| Represents no effect or no difference           | Represents an effect or difference              |
| Assumed true unless evidence suggests otherwise | Supported only if evidence against H₀ is strong |
| Contains equality (=, ≤, or ≥)                  | Contains inequality (≠, >, or <)                |
| Serves as the baseline for testing              | Represents the research claim                   |


Q3 - Explain the significance level in hypothesis testing and its role in deciding
the outcome of a test

  ANS - The significance level, denoted by α (alpha), is a threshold used in hypothesis testing to determine whether the evidence against the null hypothesis is strong enough to reject it.

What Does the Significance Level Mean?

The significance level is the maximum probability of making a Type I error, which occurs when you reject the null hypothesis (H₀) even though it is actually true.

For example:

α = 0.05 means you are willing to accept a 5% risk of incorrectly rejecting a true null hypothesis.
α = 0.01 means you are willing to accept only a 1% risk of such an error.
Role in Hypothesis Testing

After collecting data, you calculate a p-value, which indicates how likely the observed results are if H₀ is true.

The decision rule is:

If p-value ≤ α, reject H₀.
If p-value > α, fail to reject H₀.
Common Significance Levels
| Significance Level (α) | Interpretation                           |
| ---------------------- | ---------------------------------------- |
| 0.10                   | 10% risk of Type I error                 |
| 0.05                   | 5% risk of Type I error (most common)    |
| 0.01                   | 1% risk of Type I error (more stringent) |


Q4 - What are Type I and Type II errors? Give examples of each.

  ANS - In hypothesis testing, Type I and Type II errors are two kinds of mistakes that can occur when making decisions about the null hypothesis.

Type I Error (False Positive)

A Type I error occurs when you reject the null hypothesis (H₀) even though it is true.

Definition
Reality: H₀ is true.
Decision: Reject H₀.
Example

Suppose a medical test is checking for a disease.

H₀: The person does not have the disease.
H₁: The person has the disease.

A Type I error occurs if the test indicates that the person has the disease when they are actually healthy.

Consequences
Unnecessary treatment
Anxiety and additional testing
Increased costs

The probability of a Type I error is denoted by α (the significance level).

Type II Error (False Negative)

A Type II error occurs when you fail to reject the null hypothesis even though it is false.

Definition
Reality: H₀ is false.
Decision: Fail to reject H₀.
Example

Using the same medical test:

H₀: The person does not have the disease.
H₁: The person has the disease.

A Type II error occurs if the test reports that the person is healthy when they actually have the disease.

Consequences
Delayed treatment
Disease progression
Potential harm to the patient

The probability of a Type II error is denoted by β (beta)
Comparison Table
| Situation                          | Reality  | Decision          | Error Type    |
| ---------------------------------- | -------- | ----------------- | ------------- |
| Reject H₀ when H₀ is true          | H₀ true  | Reject H₀         | Type I Error  |
| Fail to reject H₀ when H₀ is false | H₀ false | Fail to reject H₀ | Type II Error |

Q6 -  Write a Python program to generate a binomial distribution with n=10 and
p=0.5, then plot its histogram.

  ANS - Here's a simple Python program that generates random samples from a binomial distribution with n=10 and p=0.5, and then plots a histogram of the generated values.

```python
import numpy as np
import matplotlib.pyplot as plt

# Parameters
n = 10      # Number of trials
p = 0.5     # Probability of success
size = 1000 # Number of samples

# Generate binomial distribution samples
data = np.random.binomial(n, p, size)

# Plot histogram
plt.hist(data, bins=np.arange(-0.5, n + 1.5, 1),
         edgecolor='black', alpha=0.7)

plt.title('Binomial Distribution (n=10, p=0.5)')
plt.xlabel('Number of Successes')
plt.ylabel('Frequency')
plt.xticks(range(n + 1))
plt.grid(axis='y', linestyle='--', alpha=0.7)

plt.show()
```

Explanation
np.random.binomial(n, p, size) generates size random observations from a binomial distribution.
bins=np.arange(-0.5, n + 1.5, 1) ensures each possible outcome (0–10) gets its own histogram bar.
plt.xticks(range(n + 1)) labels the x-axis with all possible numbers of successes.

A typical histogram will be approximately symmetric around 5, since the expected value is:

E(X)=np=10×0.5=5

and the variance is:

Var(X)=np(1−p)=10×0.5×0.5=2.5

The histogram should show the highest frequencies near 5 successes and lower frequencies toward 0 and 10.

  ANS - Let's perform a one-sample Z-test on the given dataset.

Problem Setup

Suppose we want to test whether the population mean is 50.

Null Hypothesis (H₀): μ = 50
Alternative Hypothesis (H₁): μ ≠ 50
Significance level: α = 0.05

Since a Z-test requires a known population standard deviation, we'll assume:

σ=0.6


```python
import numpy as np
from scipy.stats import norm

# Sample data
sample_data = [
    49.1, 50.2, 51.0, 48.7, 50.5, 49.8, 50.3, 50.7, 50.2, 49.6,
    50.1, 49.9, 50.8, 50.4, 48.9, 50.6, 50.0, 49.7, 50.2, 49.5,
    50.1, 50.3, 50.4, 50.5, 50.0, 50.7, 49.3, 49.8, 50.2, 50.9,
    50.3, 50.4, 50.0, 49.7, 50.5, 49.9
]

# Parameters
mu0 = 50          # Hypothesized population mean
sigma = 0.6       # Known population standard deviation
alpha = 0.05

# Sample statistics
n = len(sample_data)
x_bar = np.mean(sample_data)

# Z-statistic
z = (x_bar - mu0) / (sigma / np.sqrt(n))

# Two-tailed p-value
p_value = 2 * (1 - norm.cdf(abs(z)))

# Results
print("Sample Size:", n)
print("Sample Mean:", round(x_bar, 4))
print("Z-statistic:", round(z, 4))
print("P-value:", round(p_value, 6))

# Decision
if p_value < alpha:
    print("Reject H0")
else:
    print("Fail to Reject H0")
```

Q8 - Write a Python script to simulate data from a normal distribution and
calculate the 95% confidence interval for its mean. Plot the data using Matplotlib.

  ANS - The following Python script:

Simulates data from a normal distribution.
Calculates the sample mean and standard deviation.
Computes the 95% confidence interval for the population mean.
Plots a histogram of the simulated data with the mean and confidence interval marked


```python
import numpy as np
import matplotlib.pyplot as plt
from scipy.stats import t

# Set random seed for reproducibility
np.random.seed(42)

# Generate data from a normal distribution
mu = 50        # True mean
sigma = 10     # True standard deviation
n = 100        # Sample size

data = np.random.normal(loc=mu, scale=sigma, size=n)

# Sample statistics
sample_mean = np.mean(data)
sample_std = np.std(data, ddof=1)

# 95% Confidence Interval
confidence = 0.95
alpha = 1 - confidence

t_critical = t.ppf(1 - alpha/2, df=n-1)
margin_error = t_critical * (sample_std / np.sqrt(n))

ci_lower = sample_mean - margin_error
ci_upper = sample_mean + margin_error

# Print results
print(f"Sample Mean: {sample_mean:.2f}")
print(f"Sample Standard Deviation: {sample_std:.2f}")
print(f"95% Confidence Interval: ({ci_lower:.2f}, {ci_upper:.2f})")

# Plot histogram
plt.figure(figsize=(8, 5))
plt.hist(data, bins=12, color='skyblue', edgecolor='black')

plt.axvline(sample_mean, color='red',
            linestyle='--', linewidth=2,
            label=f'Mean = {sample_mean:.2f}')

plt.axvline(ci_lower, color='green',
            linestyle=':', linewidth=2,
            label=f'CI Lower = {ci_lower:.2f}')

plt.axvline(ci_upper, color='green',
            linestyle=':', linewidth=2,
            label=f'CI Upper = {ci_upper:.2f}')

plt.title('Normal Distribution Sample')
plt.xlabel('Value')
plt.ylabel('Frequency')
plt.legend()
plt.grid(alpha=0.3)

plt.show()
```

Q9 - Write a Python function to calculate the Z-scores from a dataset and
visualize the standardized data using a histogram. Explain what the Z-scores represent
in terms of standard deviations from the mean.

  ANS - A Z-score standardizes data by expressing each value in terms of how many standard deviations it is from the mean.

The Z-score formula is:

```python
import numpy as np
import matplotlib.pyplot as plt

def calculate_z_scores(data):
    """
    Calculate Z-scores for a dataset.
    """
    data = np.array(data)

    mean = np.mean(data)
    std = np.std(data, ddof=1)  # Sample standard deviation

    z_scores = (data - mean) / std

    return z_scores, mean, std


# Example dataset
data = [45, 50, 55, 60, 52, 48, 57, 63, 49, 51,
        54, 58, 47, 53, 61]

# Calculate Z-scores
z_scores, mean, std = calculate_z_scores(data)

print("Mean:", round(mean, 2))
print("Standard Deviation:", round(std, 2))
print("Z-Scores:")
print(np.round(z_scores, 2))

# Plot histogram of standardized data
plt.figure(figsize=(8, 5))
plt.hist(z_scores, bins=8,
         color='skyblue',
         edgecolor='black')

plt.axvline(0, color='red',
            linestyle='--',
            linewidth=2,
            label='Mean (Z = 0)')

plt.title('Histogram of Z-Scores')
plt.xlabel('Z-Score')
plt.ylabel('Frequency')
plt.legend()
plt.grid(alpha=0.3)

plt.show()
```

Interpretation of Z-Scores
| Z-Score | Meaning                              |
| ------- | ------------------------------------ |
| 0       | Exactly at the mean                  |
| +1      | 1 standard deviation above the mean  |
| +2      | 2 standard deviations above the mean |
| -1      | 1 standard deviation below the mean  |
| -2      | 2 standard deviations below the mean |


Examples
Z = 1.5 → The value is 1.5 standard deviations above the mean.
Z = -0.8 → The value is 0.8 standard deviations below the mean.
Z = 0 → The value equals the mean.
Why Use Z-Scores?
Standardization
Converts different datasets to a common scale.
Outlier Detection
Values with ∣Z∣>3 are often considered potential outliers.
Comparison Across Datasets
Allows comparison of observations measured in different units.
Empirical Rule for Normal Distributions

Approximately:

68% of observations lie within ±1σ (Z between -1 and 1)
95% lie within ±2σ (Z between -2 and 2)
99.7% lie within ±3σ (Z between -3 and 3)

Therefore, the histogram of Z-scores is centered at 0 and shows the distribution of observations relative to the dataset's mean and standard deviation.
